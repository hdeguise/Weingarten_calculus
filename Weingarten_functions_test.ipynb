```
(* make a list of all Weingarten functions for Subscript[S, enn ]*)
enn = 7;
listofclasses = 
 IntegerPartitions[
  enn]  (* list all the classes of Subscript[S, 8 ]as partitions *)
Length[listofclasses] (* number of partitions in the list; this is \
the number of distinct Weingarten functions *)
t00 = AbsoluteTime[];  (* track start time *)
lengthofclass = 
 Table[gClass[listofclasses[[qq]]], {qq, 1, Length[listofclasses]}]
tableofWg = 
  ParallelTable[
   cWg[listofclasses[[qq]], d], {qq, 1, Length[listofclasses]}];
Grid[Table[{listofclasses[[qq]], tableofWg[[qq]]}, {qq, 1, 
   Length[listofclasses]}]] 
t11 = AbsoluteTime[]; (* track end time *)
Print["This computation took ", t11 - t00, " seconds."]
```
